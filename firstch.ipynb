```python
#importing necessary libraries
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#reading and showing an image 
im_g = cv.imread("assignment1.jpg") # imread is the function used for reading an image 
plt.figure(figsize=(10, 5))
plt.imshow(im_g, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show() # this function will show the image 
cv.imshow("Image", im_g) # we can also use this function to read an image but make sure to call cv.waitKey(0) and cv.destroyAllWindows() after this 
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
def rescale_image(img,scale=0.5):
    length= int(img.shape[0]*scale)
    width = int(img.shape[1]*scale)
    dimensions =(width, length)
    return cv.resize(img, dimensions, interpolation= cv.INTER_AREA)
img= cv.imread("assignment1.jpg")
rescaled_image = rescale_image(img)
plt.figure(figsize=(10, 5))
plt.imshow(rescaled_image, cmap='gray')
plt.title("Rescaled Image")
plt.axis('off')
plt.show() 
```
