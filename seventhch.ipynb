```python
# in this we will use different blurring methods
# Blurring is an image processing technique that smooths an image by reducing sharp edges and noise
# we will use different blurring methods like average blurring, gaussian blurring and median blurring
```

```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
img1=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
#first we will use average blurring
#in avg blurring we take a pixel and take the values of its neighbouring pixels
#the avg of the neighbouring pixels values will be used in place of its own original value
avgblur_img = cv.blur(img1,(11,11))
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(avgblur_img, cmap='gray')
plt.title("AVG Blur Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# we will use gaussian blur now
# in gaussian blur we assign weight to neighbouring pixels, more weight to nearby pixels and less to distant ones
# which creats a more natural looking blur
gaussblur_img = cv.GaussianBlur(img1,(11,11),0)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(avgblur_img, cmap='gray')
plt.title("AVG Blur Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(gaussblur_img, cmap='gray')
plt.title("Gaussian Blur Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```
