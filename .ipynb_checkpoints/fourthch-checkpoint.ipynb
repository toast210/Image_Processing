```python
# In this we will be doing contour detection.
# Contour in image processing is a curve that joins all continuous points along a boundary with the same color or intensity.
# Hierarchy describes how contours are nested within each other (parent-child relationships).
# before contour detection we have to prepprocess the images using techniques like thresholding or Canny edge detection.
```

```python
# import some basic libraries required for image processing
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
plt.figure(figsize=(10, 5))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# finding the contours using canny edge detection
canny_img = cv.Canny(img,125,225)

# finding contours using find contours
# cv.RETR_EXTERNAL is one of the contour retrieval modes in OpenCV. It retrieves only the outermost contours, ignoring any nested (child) contours inside them.
# cv.CHAIN_APPROX_SIMPLE is a contour approximation method in OpenCV that reduces the number of points stored for a contour, making it more efficient.
# cv.CHAIN_APPROX_SIMPLE is used because it removes unnecessary points while keeping the shape.
contours, hierarchy = cv.findContours(canny_img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
print(f'{len(contours)} is the number of contours found')

# we can also draw the contours on a blank image to check the accuracy off contours
blank_img = np.zeros(img.shape , dtype='uint8') # blank image of same dimensions of original image
cv.drawContours(blank_img, contours, -1 , (0,255,0), 2) # drawing contours on the blank image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(canny_img, cmap='gray')
plt.title("Canny  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(blank_img, cmap='gray')
plt.title("Blank Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```
