```python
# import some basic libraries required for image processing
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# as we can see the image is different from original
img= cv.imread("assignment2.jpg")
plt.figure(figsize=(10, 5))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# converting original image in grayscaled image 
grayscaled_image=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
#with this we can see what the function is doing side by side
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(grayscaled_image, cmap='gray')
plt.title("Grayscaled Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# converting the original image to a blurred image, we use blurring to reduce the noise from the image
# we are using Gaussian blur in which each pixel is replaced by a weighted average of its neighbors
blurred_image= cv.GaussianBlur(img,(11,11),cv.BORDER_DEFAULT)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#cropping an image
cropped_img= img[400:500,600:800]
cv.imshow("Original Image", img) 
cv.imshow("Cropped Image",cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# resizing the image using resize function directly
resized_img = cv.resize(img,(300,300))
cv.imshow("Original Image", img) 
cv.imshow("Cropped Image",resized_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# edge cascading is used to find the edges in an image canny function of cv2 is used in this 
cascaded_img= cv.Canny(img,100,100)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cascaded_img, cmap='gray')
plt.title("Cascaded Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# dilating an image 
dilated_img = cv.dilate(cascaded_img,(9,9),iterations=3)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cascaded_img, cmap='gray')
plt.title("Cascaded  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(dilated_img, cmap='gray')
plt.title("Dialated Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# eroding an dilated image 
eroded_img= cv.erode(dilated_img,(9,9),iterations=3)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(dilated_img, cmap='gray')
plt.title("Dilated  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(eroded_img, cmap='gray')
plt.title("Eroded Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```
