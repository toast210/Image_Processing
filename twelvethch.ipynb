```python
# in this we will be doing edge detection
# edge detection identify the boundaries between objects in an image
# it works by detecting significant intensity changes in an image
```

```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
img1=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# convert the image to grayscale
# because it makes computation easy, and some functions only take one value
img= cv.imread("assignment2.jpg")
img2=cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 5))
plt.imshow(img2, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# using laplacian edge detection method
# in this we first fid the derivative(second-order) which finds rapid changes in intensity
lap = cv.Laplacian(img2, cv.CV_64F)  # CV_64F ensures we capture both positive and negative gradient values accurately.
lap = np.uint8(np.absolute(laplacian))  # Convert back to uint8
plt.figure(figsize=(10, 5))
plt.imshow(lap, cmap='gray')
plt.title("Laplacian Image")
plt.axis('off')
plt.show()
```

```python
# using canny edge detector
# a multi-stage algorithm that helps identify sharp changes in intensity
# multistage algorithm is a process that solves a problem by breaking it down into multiple sequential steps
# where the output of one step is used as the input for the next
edges = cv.Canny(img2, 110, 150)
plt.figure(figsize=(10, 5))
plt.imshow(edges, cmap='gray')
plt.title("Canny Image")
plt.axis('off')
plt.show()
```
