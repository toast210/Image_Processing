```python
# in this we will be doing some face detection using open cv
# opencV provides pre-trained haar models for face detection
# from this u can see the xml file used for face detection
# https://github.com/opencv/opencv/blob/4.x/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we can also do this to a grayscaled image 
img1= cv.imread("person.jpg")
img2=cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
img3=cv.cvtColor(img1,cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img2, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
haar_cascade = cv.CascadeClassifier('haar_face.xml')
```

```python
faces_detected = haar_cascade.detectMultiScale(img2, scaleFactor = 1.5, minNeighbors = 10)
print(f'No. of faces detected = {len(faces_detected)}')
for (x,y,w,h) in faces_detected:
    cv.rectangle(img3, (x,y), (x+w,y+h), (255,0,0), thickness = 20)
plt.figure(figsize=(10, 5))
plt.imshow(img3, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```
