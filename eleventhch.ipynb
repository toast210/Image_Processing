```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
img1=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
img2=cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 5))
plt.imshow(img2, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
_, thresh = cv.threshold(img2, 40, 255, cv.THRESH_BINARY)
plt.figure(figsize=(10, 5))
plt.imshow(thresh, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
plt.show()
```

```python
_, thresh = cv.threshold(img2, 40, 255, cv.THRESH_BINARY_INV)
plt.figure(figsize=(10, 5))
plt.imshow(thresh, cmap='gray')
plt.title("Inverse Binary Image")
plt.axis('off')
plt.show()
```

```python
# using adaptive thresholding method
adaptive_thresh = cv.adaptiveThreshold(img2,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,7,5)
plt.figure(figsize=(10, 5))
plt.imshow(adaptive_thresh, cmap='gray')
plt.title("Adaptive Threshold Image")
plt.axis('off')
plt.show()
```

```python
# using adaptive thresholding method
adaptive_thresh = cv.adaptiveThreshold(img2,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,7,5)
plt.figure(figsize=(10, 5))
plt.imshow(adaptive_thresh, cmap='gray')
plt.title("Adaptive Threshold Image")
plt.axis('off')
plt.show()
```
