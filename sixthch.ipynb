```python
# in this we will learn about different color channels
# as we know that there are three different parts of an image r, g ,b image
```

```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
img1=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# we will use cv.split(sorurce image) function to split the image in the repective r,g,b parts
r,g,b = cv.split(img1)
print(img.shape)
print(r.shape)
print(g.shape)
print(b.shape)
```

```python
# in this we will see the different intensities of r,g,b paets of the image
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(r, cmap='gray')
plt.title("Red Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title("Green Image")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(b, cmap='gray')
plt.title("Blue Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# we will merge the r,g,b parts and see if the resultant is same as original
rgb_merged = cv.merge([r,g,b])
plt.figure(figsize=(10, 5))
plt.imshow(rgb_merged, cmap='gray')
plt.title("Merged Image")
plt.axis('off')
plt.show()
```

```python
# we will make an blank image of same shape of original image
# merge the respective r,g,b value we got from cv.split with the blank image to get the respective values
blank_img = np.zeros(img.shape[:2],dtype='uint8')
red_img = cv.merge([r,blank_img,blank_img])
blue_img = cv.merge([blank_img,blank_img,b])
green_img = cv.merge([blank_img,g,blank_img])
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(red_img, cmap='gray')
plt.title("Red Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(green_img, cmap='gray')
plt.title("Green Image")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(blue_img, cmap='gray')
plt.title("Blue Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```
