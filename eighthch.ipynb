```python
# in this we will talk about different bitwise operators 
# there differet bitwise operators like and, or , xor, not
```

```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make some blank images
img1 = np.zeros((500,500), dtype='uint8')
plt.figure(figsize=(10, 5))
plt.imshow(img1, cmap='gray')
plt.title("Image 1")
plt.axis('off')
plt.show()
```

```python
rectangle = cv.rectangle(img1.copy(), (50,50), (450,450), 255, -1)
circle = cv.circle(img1.copy(), (250,250), 250, 255, -1)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(rectangle, cmap='gray')
plt.title("Rectangle")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(circle, cmap='gray')
plt.title("Circle")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# bitwise and - it basically places both images on each other and show only the intersection of both
bitwise_and = cv.bitwise_and(rectangle,circle)
plt.figure(figsize=(10, 5))
plt.imshow(bitwise_and, cmap='gray')
plt.title("Bitwise And")
plt.axis('off')
plt.show()
```

```python
# bitwise or - it gives the union of both images
bitwise_or = cv.bitwise_or(rectangle, circle)
plt.figure(figsize=(10, 5))
plt.imshow(bitwise_or, cmap='gray')
plt.title("Bitwise Or")
plt.axis('off')
plt.show()
```

```python
# bitwise xor - nit basically returns non intersecting part of the images
bitwise_xor = cv.bitwise_xor(rectangle, circle)
plt.figure(figsize=(10, 5))
plt.imshow(bitwise_xor, cmap='gray')
plt.title("Bitwise Xor")
plt.axis('off')
plt.show()
```

```python
bitwise_not = cv.bitwise_not(rectangle)
plt.figure(figsize=(10, 5))
plt.imshow(bitwise_not, cmap='gray')
plt.title("Bitwise Not")
plt.axis('off')
plt.show()
```

```python
bitwise_not = cv.bitwise_not(circle)
plt.figure(figsize=(10, 5))
plt.imshow(bitwise_not, cmap='gray')
plt.title("Bitwise Not")
plt.axis('off')
plt.show()
```
