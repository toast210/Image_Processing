```python
# import necessary liberaries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
img1=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img1, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# splitting the image in its components
r,g,b = cv.split(img1)
```

```python
# visualization of the histograms 
plt.figure(figsize=(10, 2))
plt.subplot(1, 3, 1)
plt.hist(r.ravel(), 256, [0, 256], color="red")
plt.title("RED ")
plt.axis()
plt.subplot(1, 3, 2)
plt.hist(g.ravel(), 256, [0, 256], color="green")
plt.title("GREEN ")
plt.axis()
plt.subplot(1, 3, 3)
plt.hist(b.ravel(), 256, [0, 256], color="blue")
plt.title("BLUE ")
plt.axis()
plt.tight_layout()
plt.show()
```

```python
# using histogram equalization
r_eq=cv.equalizeHist(r)
g_eq=cv.equalizeHist(g)
b_eq=cv.equalizeHist(b)
```

```python
# visualizing the equalized r,g,b components
plt.figure(figsize=(10, 2))
plt.subplot(1, 3, 1)
plt.hist(r_eq.ravel(), 256, [0, 256], color="red")
plt.title("RED Equalized")
plt.axis()
plt.subplot(1, 3, 2)
plt.hist(g_eq.ravel(), 256, [0, 256], color="green")
plt.title("GREEN Equalized")
plt.axis()
plt.subplot(1, 3, 3)
plt.hist(b_eq.ravel(), 256, [0, 256], color="blue")
plt.title("BLUE Equalized")
plt.axis()
plt.tight_layout()
plt.show()
```

```python
# merging the eqalized components
equalized_img = cv.merge([r_eq, g_eq, b_eq])
plt.figure(figsize=(10, 5))
plt.imshow(equalized_img, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')
plt.show()
```
