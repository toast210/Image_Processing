```python
# import some basic libraries required for image processing
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv.imread("assignment2.jpg")
plt.figure(figsize=(10, 5))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# translating an image 
# function for moving/translating an image
def move_img(img,x,y):
    movemat= np.float32([[1,0,x],[0,1,y]])
    dimensions=(img.shape[1],img.shape[0])
    return cv.warpAffine(img,movemat,dimensions)
# -x --> left
# -y --> up
#  x --> right
#  y --> down
x=int(input("Move image left(-ve) and right(+ve)"))
y=int(input("Move image up(-ve) and down(+ve)"))
moved_img=move_img(img,x,y)

# showing comparison between images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(moved_img, cmap='gray')
plt.title("Translated Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# rotating an image by some angle
# making a function to rotate the image on its center
def rotate_img(img, angle, point=None):
    (height,width)= img.shape[:2]
    if point is None:
        point = (width//2,height//2)
    rotation_matrix=cv.getRotationMatrix2D(point, angle, 1.0)
    dimensions=(width,height)
    return cv.warpAffine(img,rotation_matrix,dimensions)
# +ve for anti-clockwise
# -ve for clockwise
rotated_img = rotate_img(img,180)
# showing comparison between images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(rotated_img, cmap='gray')
plt.title("Rotated Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# flipping an image 
# 0 to flip image vetically
# 1 to flip image horizontally
# -1 to flip image both horizontally ang vetically
flipped_img = cv.flip(img,1)
# showing comparison between images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original  Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(flipped_img, cmap='gray')
plt.title("Flipped Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```
